## Time Resampling

```python
data_tesla.head()
```

```python
data_tesla.resample(rule='A').max()
```

```python
data_tesla.resample(rule='A').min()
```

```python
data_tesla.resample(rule='A').max()['Open'].plot()

##search for different rules on internet
##Rule A is for year end frequency
```

```python
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
```

```python
##Quaterly
data_tesla.resample(rule='QS').max()
```

```python
plt.figure(figsize=(20,4))
ax = data_tesla.resample(rule='QS').max()['Open'].plot()
ax.set_xticks(data_tesla.index)
ax.set_xticklabels([date.strftime('%Y-%m-%d') for date in data_tesla.index], rotation=45)

plt.xlabel('Date')
plt.ylabel('Open Price')
plt.title('Quarterly Maximum Open Price for Tesla')
plt.show()
```

```python
data_tesla.resample(rule='BA').max()
```

```python
data_tesla.resample(rule='BA').max()['Open'].plot()
```

```python
data_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
data_tesla_monthly = data_tesla.resample(rule='M').mean()
data_tesla_monthly_2020 = data_tesla_monthly[data_tesla_monthly.index >= '2020-01-01']

data_tesla_monthly_2020['Open'].plot(kind='bar',figsize=(20, 4))
plt.xlabel('Date')
plt.ylabel('Mean Open Price')
plt.title('Monthly Mean Open Price for Tesla Starting from 2020')
```
